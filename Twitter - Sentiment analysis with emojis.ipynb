```python
from spacy.lang.en import English
from spacy.matcher import Matcher

nlp = English()
matcher = Matcher(nlp.vocab)

information_emoji = ["猬锔","","","","","","","","","",""]
ideologia_emoji = ["","","斥","",""]
positivo_emoji = ["", "", "","","",""]
aprobacion_emoji = ["","","","","","","わ","","",""]
desacuerdo_emoji = ["","","","","","あ","う","","し","",""]
risa_emoji = ["","ぃ",""]
tristeza_emoji = ["",  "", "", ""]



# Se le pasan los patrones positivos de los emojis

information_patterns = [[{"ORTH": emoji}] for emoji in information_emoji]
ideologia_patterns = [[{"ORTH": emoji}] for emoji in ideologia_emoji]
positivo_patterns = [[{"ORTH": emoji}] for emoji in positivo_emoji]
aprobacion_patterns = [[{"ORTH": emoji}] for emoji in aprobacion_emoji]
desacuerdo_patterns = [[{"ORTH": emoji}] for emoji in desacuerdo_emoji]
risa_emoji_patterns = [[{"ORTH": emoji}] for emoji in risa_emoji]
tristeza_patterns = [[{"ORTH": emoji}] for emoji in tristeza_emoji]



# Etiquetamos el sentimiento a trav茅s de encontrar emojis
def label_sentiment(matcher, doc, i, matches):
    global cuentainformation
    global cuentaideologia
    global cuentapositivo
    global cuentaaprobacion
    global cuentadesacuerdo
    global cuentarisa
    global cuentatristeza
    
    match_id, start, end = matches[i]
    if doc.vocab.strings[match_id] == "INFO":
        cuentainformation += 1  # aumentamos 0.1 por el sentimiento positivo
    elif doc.vocab.strings[match_id] == "IDE":
        cuentaideologia += 1  # restamos 0.1 por el sentimiento negativo
    elif doc.vocab.strings[match_id] == "POS":
        cuentapositivo += 1  # restamos 0.1 por el sentimiento negativo
    elif doc.vocab.strings[match_id] == "APR":
        cuentaaprobacion += 1  # restamos 0.1 por el sentimiento negativo        
    elif doc.vocab.strings[match_id] == "DES":
        cuentadesacuerdo += 1  # restamos 0.1 por el sentimiento negativo        
    elif doc.vocab.strings[match_id] == "RIS":
        cuentarisa += 1  # restamos 0.1 por el sentimiento negativo
    elif doc.vocab.strings[match_id] == "TRI":
        cuentatristeza += 1  # restamos 0.1 por el sentimiento negativo        
    elif doc.vocab.strings[match_id] == "HASHTAG":
        cuentahashtags += 1  # restamos 0.1 por el sentimiento negativo           
        
matcher.add("INFO", information_patterns, on_match=label_sentiment)  
matcher.add("IDE", ideologia_patterns, on_match=label_sentiment)  
matcher.add("POS", positivo_patterns, on_match=label_sentiment)  
matcher.add("APR", aprobacion_patterns, on_match=label_sentiment)
matcher.add("DES", desacuerdo_patterns, on_match=label_sentiment) 
matcher.add("RIS", risa_emoji_patterns, on_match=label_sentiment)  
matcher.add("TRI", tristeza_patterns, on_match=label_sentiment)


# A帽adimos el patr贸n Hashtag, '#' y que sea de tipo ascii
matcher.add("HASHTAG", [[{"ORTH": "#"}, {"IS_ASCII": True}]])


for politico in politicos:
    cuentainformation = 0
    cuentaideologia = 0
    cuentapositivo = 0
    cuentaaprobacion = 0
    cuentadesacuerdo = 0
    cuentarisa = 0
    cuentatristeza = 0
    cuentahashtags = 0
    input("vamos con: "+politico)
    for tweet in tweets[politico]:
        doc = nlp(tweet)
        matches = matcher(doc)
        nt=len(tweets[politico])
        ct = cuentainformation + cuentaideologia + cuentapositivo + cuentaaprobacion + cuentadesacuerdo + cuentarisa + cuentatristeza + cuentahashtags
        for match_id, start, end in matches:
            string_id = doc.vocab.strings[match_id]
            span = doc[start:end]
            #print(string_id, span.text)
    print("\n""% Iconos de informaci贸n: ",str(int(cuentainformation/ct*100))," %","\n""% Iconos de IDEOLOGA: "
          ,str(int(cuentaideologia/ct*100))," %","\n""% Iconos POSITIVOS: ",str(int(cuentapositivo/ct*100))," %","\n""ratio iconos de aprobaci贸n: "
          ,str(int(cuentaaprobacion/ct*100))," %","\n""ratio iconos de desacuerdo: ",str(int(cuentadesacuerdo/ct*100))," %","\n""ratio iconos de risa: "
          ,str(int(cuentarisa/ct*100))," %","\n""ratio iconos de tristeza: ",str(int(cuentatristeza/ct*100))," %","\n\ntotal tweets: ",nt,"\n"
          )
```
